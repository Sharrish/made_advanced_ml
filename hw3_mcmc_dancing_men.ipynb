```python
import re
from collections import Counter

import random
random.seed(1234)

import copy

import numpy as np

# –†—É—Å—Å–∫–∏–π –∞–ª—Ñ–∞–≤–∏—Ç —Å –ø—Ä–æ–±–µ–ª–æ–º, –Ω–æ –±–µ–∑ –±—É–∫–≤—ã —ë.
RUSSIAN_ALPHABET_SPACE = "–∞–±–≤–≥–¥–µ–∂–∑–∏–π–∫–ª–º–Ω–æ–ø—Ä—Å—Ç—É—Ñ—Ö—Ü—á—à—â—ä—ã—å—ç—é—è "


# --- –î–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏ --- 
import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns
sns.set_style("whitegrid")
sns.set_palette("colorblind")
palette = sns.color_palette()

plt.style.use(['science','ieee','no-latex']) # –¥–ª—è —Ñ–æ—Ä–º—É–ª –≤ –≥—Ä–∞—Ñ–∏–∫–∞—Ö
plt.rc('font',**{'family':'sans-serif'}) # —à—Ä–∏—Ñ—Ç
plt.rc('figure', **{'dpi': 1000}) # –∫–∞—á–µ—Å—Ç–≤–æ –≥—Ä–∞—Ñ–∏–∫–æ–≤


figsize = (15, 8)
legend_fontsize = 19
axis_fontsize = 16
# --- ------------- --- 
```

## 1Ô∏è‚É£ –ë–∞–∑–æ–≤—ã–π —á–∞—Å—Ç–æ—Ç–Ω—ã–π –º–µ—Ç–æ–¥ –ø–æ –®–µ—Ä–ª–æ–∫—É –•–æ–ª–º—Å—É

–†–µ–∞–ª–∏–∑—É–π—Ç–µ –±–∞–∑–æ–≤—ã–π —á–∞—Å—Ç–æ—Ç–Ω—ã–π –º–µ—Ç–æ–¥ –ø–æ –®–µ—Ä–ª–æ–∫—É –•–æ–ª–º—Å—É:
* –ø–æ–¥—Å—á–∏—Ç–∞–π—Ç–µ —á–∞—Å—Ç–æ—Ç—ã –±—É–∫–≤ –ø–æ –∫–æ—Ä–ø—É—Å–∞–º (–ø—É–Ω–∫—Ç—É–∞—Ü–∏—é –∏ –∫–∞–ø–∏—Ç–∞–ª–∏–∑–∞—Ü–∏—é –º–æ–∂–Ω–æ –ø—Ä–æ—Å—Ç–æ –æ–ø—É—Å—Ç–∏—Ç—å, –∞ –≤–æ—Ç –ø—Ä–æ–±–µ–ª—ã –ª—É—á—à–µ –æ—Å—Ç–∞–≤–∏—Ç—å);
* –≤–æ–∑—å–º–∏—Ç–µ –∫–∞–∫–∏–µ-–Ω–∏–±—É–¥—å —Ç–µ—Å—Ç–æ–≤—ã–µ —Ç–µ–∫—Å—Ç—ã (–Ω—É–∂–Ω–æ –≤–∑—è—Ç—å –ø–æ –º–µ–Ω—å—à–µ–π –º–µ—Ä–µ $2$-$3$ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è, –∏–Ω–∞—á–µ –≤—Ä—è–¥ –ª–∏ —Å—Ä–∞–±–æ—Ç–∞–µ—Ç), –∑–∞—à–∏—Ñ—Ä—É–π—Ç–µ –∏—Ö –ø–æ—Å—Ä–µ–¥—Å—Ç–≤–æ–º —Å–ª—É—á–∞–π–Ω–æ–π –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ —Å–∏–º–≤–æ–ª–æ–≤;
* —Ä–∞—Å—à–∏—Ñ—Ä—É–π—Ç–µ –∏—Ö —Ç–∞–∫–∏–º —á–∞—Å—Ç–æ—Ç–Ω—ã–º –º–µ—Ç–æ–¥–æ–º.  

```python
# –ß–∏—Ç–∞–µ–º "–í–æ–π–Ω—É –∏ –º–∏—Ä" –ø–æ-—Ä—É—Å—Å–∫–∏.
with open("data/WarAndPeaceRus.txt", "r", encoding="utf-8") as fin:
    corpus_rus_lines = fin.readlines()
    
# –ö–æ–Ω–∫–∞–Ω—Ç–µ–Ω–∏—Ä—É–µ–º –≤—Å–µ —Å—Ç—Ä–æ–∫–∏ –≤ –µ–¥–∏–Ω—É—é —Å—Ç—Ä–æ–∫—É.
corpus_rus = ''.join(corpus_rus_lines)
```

```python
def preprocess_rus_text(text: str) -> str:
    """–ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ —Ä—É—Å—Å–∫–æ–≥–æ —Ç–µ–∫—Å—Ç–∞."""
    # –ö–∞–ø–∏—Ç–∞–ª–∏–∑–∞—Ü–∏—é –º–æ–∂–Ω–æ –ø—Ä–æ—Å—Ç–æ –æ–ø—É—Å—Ç–∏—Ç—å.
    text = text.lower()
    # –£–¥–∞–ª–∏–º –≤—Å–µ —Å–∏–º–≤–æ–ª—ã –∫—Ä–æ–º–µ –±—É–∫–≤ –∏ –ø—Ä–æ–±–µ–ª–∞.
    text = re.sub("[^–∞-—è ]", " ", text)
    # –£–¥–∞–ª–µ–Ω–∏–µ –¥–≤–æ–π–Ω—ã—Ö –ø—Ä–æ–±–µ–ª–æ–≤.
    text = re.sub(' +',' ', text).strip()
    return text
```

```python
print(f"–í—Å–µ–≥–æ —Å–∏–º–≤–æ–ª–æ–≤ –≤ —Ä—É—Å—Å–∫–æ–π –≤–µ—Ä—Å–∏–∏ –±—ã–ª–æ –¥–æ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏: {len(corpus_rus)}")
corpus_rus = preprocess_rus_text(corpus_rus)
print(f"–í—Å–µ–≥–æ —Å–∏–º–≤–æ–ª–æ–≤ –≤ —Ä—É—Å—Å–∫–æ–π –≤–µ—Ä—Å–∏–∏ —Å—Ç–∞–ª–æ –ø–æ—Å–ª–µ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏: {len(corpus_rus)}")
```

```python
# C—á–∏—Ç–∞–µ–º —á–∞—Å—Ç–æ—Ç—ã –±—É–∫–≤ –ø–æ –∫–æ—Ä–ø—É—Å—É.
frequencies_dict = dict(Counter(corpus_rus))
```

```python
# –°—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ —á–∞—Å—Ç–æ—Ç –±—É–∫–≤.

frequencies_list_rus = []
for symbol in RUSSIAN_ALPHABET_SPACE:
    frequencies_list_rus.append(frequencies_dict[symbol])
        
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=figsize)

ax.bar(list(RUSSIAN_ALPHABET_SPACE), frequencies_list_rus, color='dodgerblue')
ax.set_title(r'–ß–∞—Å—Ç–æ—Ç—ã —Å–∏–º–≤–æ–ª–æ–≤ –≤ "–í–æ–π–Ω–µ –∏ –º–∏—Ä–µ"', fontsize=legend_fontsize)
ax.set_xlabel(r'–°–∏–º–≤–æ–ª—ã', fontsize=axis_fontsize)
ax.set_ylabel(r'–ß–∞—Å—Ç–æ—Ç–∞ —Å–∏–º–≤–æ–ª–æ–≤', fontsize=axis_fontsize)

plt.show()
```

```python
# –®–∏—Ñ—Ä –Ω–∞ –æ—Å–Ω–æ–≤–µ —Å–ª—É—á–∞–π–Ω–æ–π –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ —Å–∏–º–≤–æ–ª–æ–≤ –∞–ª—Ñ–∞–≤–∏—Ç–∞.
SHUFFLE_RUSSIAN_ALPHABET_SPACE = list(RUSSIAN_ALPHABET_SPACE)
random.shuffle(SHUFFLE_RUSSIAN_ALPHABET_SPACE)
SHUFFLE_RUSSIAN_ALPHABET_SPACE = "".join(SHUFFLE_RUSSIAN_ALPHABET_SPACE)

print("–û—Ä–∏–≥–∏–Ω–∞–ª—å–Ω—ã–π —Ä—É—Å—Å–∫–∏–π –∞–ª—Ñ–∞–≤–∏—Ç: {}".format(RUSSIAN_ALPHABET_SPACE))
print("–®–∏—Ñ—Ä —Ä—É—Å—Å–∫–æ–≥–æ –∞–ª—Ñ–∞–≤–∏—Ç–∞: {}".format(SHUFFLE_RUSSIAN_ALPHABET_SPACE))
```

```python
# –ü–æ–ª—É—á–∞–µ–º —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è.
dict_translation = str.maketrans(RUSSIAN_ALPHABET_SPACE, SHUFFLE_RUSSIAN_ALPHABET_SPACE)
```

```python
def encode_text(text: str, dict_translation: dict) -> str:
    """–ö–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ —Ç–µ–∫—Å—Ç–∞ –Ω–∞ –æ—Å–Ω–æ–≤–µ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ —Å–∏–º–≤–æ–ª–æ–≤ –∞–ª—Ñ–∞–≤–∏—Ç–∞."""
    return text.translate(dict_translation)
```

```python
def decode_text(text: str, frequencies_dict: dict) -> str:
    """–î–µ–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ —Ç–µ–∫—Å—Ç–∞ –Ω–∞ –æ—Å–Ω–æ–≤–µ —á–∞—Å—Ç–æ—Ç–Ω–æ–≥–æ –º–µ—Ç–æ–¥–∞ –ø–æ –®–µ—Ä–ª–æ–∫—É –•–æ–ª–º—Å—É."""
    
    sorted_frequencies_corpus = sorted(list(frequencies_dict.items()), key=lambda x: x[1], reverse=True)
    frequencies_alphabet1 = "".join([x[0] for x in sorted_frequencies_corpus])
    
    frequencies_dict2 = dict(Counter(text))
    sorted_frequencies_corpus2 = sorted(list(frequencies_dict2.items()), key=lambda x: x[1], reverse=True)
    frequencies_alphabet2 = "".join([x[0] for x in sorted_frequencies_corpus2])
    
    dict_translation = str.maketrans(frequencies_alphabet2, frequencies_alphabet1)
    return text.translate(dict_translation)
```

```python
def get_accuracy_decode(source_text: str, decode_text: str) -> float:
    hit = 0
    for x1, x2 in zip(source_text, decode_text):
        if x1 == x2:
            hit += 1
    return hit / len(source_text)
```

```python
# –ß–∏—Ç–∞–µ–º "–ê–Ω–Ω—É –ö–∞—Ä–µ–Ω–∏–Ω—É" –ø–æ-—Ä—É—Å—Å–∫–∏.
with open("data/AnnaKareninaRus.txt", "r", encoding="utf-8") as fin:
    anna_rus_lines = fin.readlines()
    
# –ö–æ–Ω–∫–∞–Ω—Ç–µ–Ω–∏—Ä—É–µ–º –≤—Å–µ —Å—Ç—Ä–æ–∫–∏ –≤ –µ–¥–∏–Ω—É—é —Å—Ç—Ä–æ–∫—É.
anna_rus_text = ''.join(anna_rus_lines)
# –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ —Ç–µ–∫—Å—Ç–∞.
anna_rus_text = preprocess_rus_text(anna_rus_text)

EXAMPLE_TEXT_ANNA = anna_rus_text[32793:34793]
```

```python
def get_frequencies_bigram_dict(text: str):
    frequencies_bigram_dict = Counter([text[i:i+2] for i in range(len(text) - 1)])
    
    # –î–ª—è —Å–≥–ª–∞–∂–∏–≤–∞–Ω–∏—è –∏ —á—Ç–æ–±—ã —Å–ª–æ–≤–∞—Ä–∏ —á–∞—Å—Ç–æ—Ç –±—ã–ª–∏ –æ–¥–Ω–æ–π –¥–ª–∏–Ω—ã —É —Ä–∞–∑–Ω—ã—Ö —Ç–µ–∫—Å—Ç–æ–≤.
    for c1 in RUSSIAN_ALPHABET_SPACE:
        for c2 in RUSSIAN_ALPHABET_SPACE:
            frequencies_bigram_dict[c1 + c2] += 1
            
    return dict(frequencies_bigram_dict)
```

## 3Ô∏è‚É£ –ß–∞—Å—Ç–æ—Ç–Ω—ã–π –ø–æ–¥—Ö–æ–¥ –Ω–∞ –æ—Å–Ω–æ–≤–µ –±–∏–≥—Ä–∞–º–º
–ù–æ –∏ —ç—Ç–æ –µ—â—ë –Ω–µ –≤—Å—ë: –±–∏–≥—Ä–∞–º–º—ã —Å–∫–æ—Ä–µ–µ –≤—Å–µ–≥–æ —Ç–æ–∂–µ –¥–∞–ª–µ–∫–æ –Ω–µ –≤—Å–µ–≥–¥–∞ —Ä–∞–±–æ—Ç–∞—é—Ç. –û—Å–Ω–æ–≤–Ω–∞—è —á–∞—Å—Ç—å –∑–∞–¥–∞–Ω–∏—è ‚Äî –≤ —Ç–æ–º, –∫–∞–∫ –º–æ–∂–Ω–æ –∏—Ö —É–ª—É—á—à–∏—Ç—å:
* –ø—Ä–µ–¥–ª–æ–∂–∏—Ç–µ –º–µ—Ç–æ–¥ –æ–±—É—á–µ–Ω–∏—è –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ —Å–∏–º–≤–æ–ª–æ–≤ –≤ —ç—Ç–æ–º –∑–∞–¥–∞–Ω–∏–∏, –æ—Å–Ω–æ–≤–∞–Ω–Ω—ã–π –Ω–∞ MCMC-—Å—ç–º–ø–ª–∏—Ä–æ–≤–∞–Ω–∏–∏, –Ω–æ –ø–æ-–ø—Ä–µ–∂–Ω–µ–º—É —Ä–∞–±–æ—Ç–∞—é—â–∏–π –Ω–∞ –æ—Å–Ω–æ–≤–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –±–∏–≥—Ä–∞–º–º;
* —Ä–µ–∞–ª–∏–∑—É–π—Ç–µ –∏ –ø—Ä–æ—Ç–µ—Å—Ç–∏—Ä—É–π—Ç–µ –µ–≥–æ, —É–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —É–ª—É—á—à–∏–ª–∏—Å—å.  

<font color='Navy'><i> 
‚úèÔ∏è - –ò—Ç–∞–∫, —É –Ω–∞—Å –≤ —Ä–∞—Å–ø–æ—Ä—è–∂–µ–Ω–∏–∏ –µ—Å—Ç—å —Å–ª–æ–≤–∞—Ä—å —á–∞—Å—Ç–æ—Ç, –∫–æ—Ç–æ—Ä—É—é –º—ã –ø–æ–ª—É—á–∏–ª–∏ –ø—Ä–∏ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–∫–µ –Ω–∞ "–í–æ–π–Ω–µ –∏ –ú–∏—Ä".  
    - –ú—ã –º–æ–∂–µ–º —Ä–∞—Å—Å–º–æ—Ç—Ä–µ—Ç—å —ç—Ç–æ—Ç —Å–ª–æ–≤–∞—Ä—å –±–∏–≥—Ä–∞–º–º $s_i$, $s_{i+1}$ —á–∞—Å—Ç–æ—Ç –∫–∞–∫ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ $p(s_{i+1} \mid s_i)$ (–¥–ª—è —Å—Ç—Ä–æ–≥–∏—Ö –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –Ω–∞–¥–æ –æ—Ç–Ω–æ—Ä–º–∏—Ä–æ–≤–∞—Ç—å –Ω–∞ —Å—É–º–º—É —á–∞—Å—Ç–æ—Ç –≤—Å–µ—Ö –±–∏–≥—Ä–∞–º, –Ω–æ —ç—Ç–æ –Ω–µ –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ).  
    - –¢–æ–≥–¥–∞ –¥–ª—è –¥–∞–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–∏ $s$ –º–æ–∂–Ω–æ –≤—ã—á–∏—Å–ª–∏—Ç—å –µ—ë –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å: $\mathbb{P_1}(s) = \displaystyle\prod_i^{len(s)}{p(s_{i+1} \mid s_i)} \sim \displaystyle\prod_i^{len(s)} frequency(s_i\,s_{i+1})$.  
    - –ù–∞—à–∞ —Ü–µ–ª—å - —ç—Ç–æ –Ω–∞–π—Ç–∏ —Å–æ–æ—Ç–≤–µ—Ç—Å–≤–∏–µ –º–µ–∂–¥—É —à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω—ã–º (–ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ—á–Ω—ã–º) –∞–ª—Ñ–∞–≤–∏—Ç–æ–º –∏ –∏—Å—Ö–æ–¥–Ω—ã–º –∞–ª—Ñ–∞–≤–∏—Ç–æ–º.  
    - –í –∫–∞—á–µ—Å—Ç–≤–µ –Ω–∞—á–∞–ª—å–Ω–æ–π –∏–Ω–∏—Ü–∏–∞–ª–∏–∑—Ü–∏–∏ —ç—Ç–æ–≥–æ —Å–æ–æ—Ç–≤–µ—Ç—Å–≤–∏—è (–ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏) –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ø–æ–¥—Ö–æ–¥ –∏–∑ –ø–µ—Ä–≤–æ–≥–æ –ø—É–Ω–∫—Ç–∞ (–®–µ—Ä–ª–æ–∫–∞ –•–æ–ª–º—Å–∞).  
    - –î–∞–ª–µ–µ –º—ã –±—É–¥–µ–º –ø—ã—Ç–∞—Ç—å—Å—è –º–µ–Ω—è—Ç—å –º–µ—Å—Ç–∞–º–∏ –¥–≤–µ –ø–æ–∑–∏—Ü–∏–∏ –≤ —ç—Ç–æ–º —Å–æ–æ—Ç–≤–µ—Ç—Å–≤–∏–∏ (–ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏) –∏ —Å—á–∏—Ç–∞—Ç—å –Ω–æ–≤—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å $\mathbb{P_2}(s)$.  
    - –î–ª—è –ø—Ä–∏–Ω—è—Ç–∏—è —Ä–µ—à–µ–Ω–∏—è, —Å—Ç–æ–∏—Ç –ª–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞—Ç—å –Ω–æ–≤–æ–µ —Å–æ–æ—Ç–≤–µ—Ç—Å–≤–∏–µ –∏–ª–∏ –æ—Å—Ç–∞–≤–∞—Ç—å—Å—è –ø—Ä–∏ —Å—Ç–∞—Ä–æ–º –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ø–æ–¥—Ö–æ–¥ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –ú–µ—Ç—Ä–æ–ø–æ–ª–∏—Å–∞‚Äî–ì–∞—Å—Ç–∏–Ω–≥—Å–∞.  
    - –ï—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω–æ, —á—Ç–æ –º–æ–∂–Ω–æ –ø–µ—Ä–µ–π—Ç–∏ –æ—Ç –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–π –∫ —Å—É–º–º–∞–º —Å –ø–æ–º–æ—â—å—é –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä–æ–≤–∞–Ω–∏—è.
</i></font>

```python
def get_dict_translation(text: str, frequencies_dict: dict) -> str:
    """–ü–æ–ª—É—á–µ–Ω–∏–µ —Å–ª–æ–≤–∞—Ä—è-—Å–æ–æ—Ç–≤–µ—Ç—Å–≤–∏—è –∞–ª—Ñ–∞–≤–∏—Ç–æ–≤ –¥–ª—è –¥–µ—à–∏—Ñ—Ä–æ–≤–∫–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ –º–µ—Ç–æ–¥–∞ –®–µ—Ä–ª–æ–∫–∞ –•–æ–ª–º—Å–∞."""
    
    sorted_frequencies_corpus = sorted(list(frequencies_dict.items()), key=lambda x: x[1], reverse=True)
    frequencies_alphabet1 = "".join([x[0] for x in sorted_frequencies_corpus])
    
    frequencies_dict2 = dict(Counter(text))
    sorted_frequencies_corpus2 = sorted(list(frequencies_dict2.items()), key=lambda x: x[1], reverse=True)
    frequencies_alphabet2 = "".join([x[0] for x in sorted_frequencies_corpus2])

    #     dict_translation = str.maketrans(frequencies_alphabet2, frequencies_alphabet1)
    dict_translation = dict(zip(frequencies_alphabet2, frequencies_alphabet1))

    return dict_translation
```

```python
def get_new_dict_translation(dict_translation: dict) -> dict:
    """–í–æ–∑–≤—Ä–∞—â–∞–µ–º —Å–ª–æ–≤–∞—Ä—å —Å–æ–æ—Ç–≤–µ—Ç—Å–≤–∏–π –∞–ª—Ñ–∞–≤–∏—Ç–æ–≤, –≤ –∫–æ—Ç–æ—Ä–æ–º —Å–ª—É—á–∞–π–Ω–æ –∑–∞–º–µ–Ω–∏–ª–∏ –¥–≤–µ –ø–æ–∑–∏—Ü–∏–∏."""
    new_dict = copy.copy(dict_translation)
    symbol1, symbol2 = np.random.choice(list(new_dict.keys()), size=2, replace=False)
    new_dict[symbol1], new_dict[symbol2] = new_dict[symbol2], new_dict[symbol1]
    
    return new_dict
    
def get_log_likelihood(text: str, dict_translation: dict, frequencies_bigram_dict: dict) -> float:
    """–í–æ–∑–≤—Ä–∞—â–∞–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å—Ç—Ä–æ–∫–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ —á–∞—Å—Ç–æ—Ç –±–∏–≥—Ä–∞–º–º."""
    decode_text = translate_text(text, dict_translation) # text.translate(dict_translation)
    log_likelihood = 0
    for i in range(len(decode_text) - 2):
        log_likelihood += np.log(frequencies_bigram_dict[decode_text[i:i+2]])
    return log_likelihood


def metropolis_hastings_log_accept(l: float, l_new: float) -> bool:
    """–ü—Ä–∏–Ω–∏–º–∞–µ–º –ª–∏ –∏–∑–º–µ–Ω–µ–Ω–∏–µ –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å–≤–∏–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ –Ω–æ–≤–æ–≥–æ –ª–æ–≥–∞—Ä–∏—Ñ–º–∞ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è —Ç–µ–∫—Å—Ç–∞."""
    if l_new > l:
        return True
    else:
        return (np.random.rand() < (np.exp(l_new-l)))
    
def translate_text(text: str, dict_translation: dict) -> str:
    """–î–µ—à–∏—Ñ—Ä–æ–≤—ã–≤–∞–µ—Ç —Ç–µ—Å—Ç –Ω–∞ –æ—Å–Ω–æ–≤–µ —Å–ª–æ–≤–∞—Ä—è-—Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏—è –∞–ª—Ñ–∞–≤–∏—Ç–æ–≤."""
    return "".join([dict_translation[x] for x in text]) 
```

```python
def mcmc_decode_text(text: str, iterations: int, frequencies_dict: dict, frequencies_bigram_dict: dict) -> str:
    """–ú–µ—Ç–æ–¥ –æ–±—É—á–µ–Ω–∏—è —Å–æ–æ—Ç–≤–µ—Ç—Å–≤–∏—è –º–µ–∂–¥—É –∞–ª—Ñ–∞–≤–∏—Ç–∞–º–∏, –æ—Å–Ω–æ–≤–∞–Ω–Ω—ã–π –Ω–∞ MCMC-—Å—ç–º–ø–ª–∏—Ä–æ–≤–∞–Ω–∏–∏."""
    # –ù–∞—á–∞–ª—å–Ω–∞—è –∏–Ω–∏—Ü–∏–∞–ª–∏–∑—Ü–∏–∏ —Å–æ–æ—Ç–≤–µ—Ç—Å–≤–∏—è (–ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏) - –ø–æ–¥—Ö–æ–¥ –∏–∑ –ø–µ—Ä–≤–æ–≥–æ –ø—É–Ω–∫—Ç–∞ (–®–µ—Ä–ª–æ–∫–∞ –•–æ–ª–º—Å–∞).
    curr_dict_translation = get_dict_translation(text, frequencies_dict)
    curr_l = get_log_likelihood(text, curr_dict_translation, frequencies_bigram_dict)
    for i in range(iterations):
        new_dict_translation = get_new_dict_translation(curr_dict_translation)
        new_l = get_log_likelihood(text, new_dict_translation, frequencies_bigram_dict)
        if metropolis_hastings_log_accept(curr_l, new_l):
            curr_dict_translation, curr_l = new_dict_translation, new_l
    return translate_text(text, curr_dict_translation), curr_l  # text.translate(dict_translation)
```
